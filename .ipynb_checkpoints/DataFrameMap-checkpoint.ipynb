```python
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import pandas as pd


# Import the Shape
fp = "/home/kevin/Work/mapper/México_Estados/México_Estados.shp"
basemap = gpd.read_file(fp)
basemap.head()

# Read de Data
df = pd.read_csv("mapa_heat.csv", delimiter=',', skiprows=0, low_memory=False)

# Get Dataframes by region
df_red = df[df['golden'] <= 25]
df_yellow = df[(df['golden'] >=26) & (df['golden'] <=74) ]
df_green  = df[df['golden'] >= 75 ]


#DataFrames RED
df_red_1 = df[(df['region'] == 1.0) & (df['golden'] <= 25)]
df_red_2 = df[(df['region'] == 2.0) & (df['golden'] <= 25)]
df_red_3 = df[(df['region'] == 3.0) & (df['golden'] <= 25)]
df_red_4 = df[(df['region'] == 4.0) & (df['golden'] <= 25)] 
df_red_5 = df[(df['region'] == 5.0) & (df['golden'] <= 25)] 
df_red_6 = df[(df['region'] == 6.0) & (df['golden'] <= 25)] 
df_red_7 = df[(df['region'] == 7.0) & (df['golden'] <= 25)] 
df_red_8 = df[(df['region'] == 8.0) & (df['golden'] <= 25)] 
df_red_9 = df[(df['region'] == 9.0) & (df['golden'] <= 25)] 


#DataFrames Yellow
df_yellow_1 = df[(df['region'] == 1.0) & (df['golden'] >= 26) & (df['golden'] <= 74)]
df_yellow_2 = df[(df['region'] == 2.0) & (df['golden'] >= 26) & (df['golden'] <= 74)]
df_yellow_3 = df[(df['region'] == 3.0) & (df['golden'] >= 26) & (df['golden'] <= 74)]
df_yellow_4 = df[(df['region'] == 4.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 
df_yellow_5 = df[(df['region'] == 5.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 
df_yellow_6 = df[(df['region'] == 6.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 
df_yellow_7 = df[(df['region'] == 7.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 
df_yellow_8 = df[(df['region'] == 8.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 
df_yellow_9 = df[(df['region'] == 9.0) & (df['golden'] >= 26) & (df['golden'] <= 74)] 

#DataFrames Green
df_green_1 = df[(df['region'] == 1.0) & (df['golden'] >= 75)]
df_green_2 = df[(df['region'] == 2.0) & (df['golden'] >= 75)]
df_green_3 = df[(df['region'] == 3.0) & (df['golden'] >= 75)]
df_green_4 = df[(df['region'] == 4.0) & (df['golden'] >= 75)] 
df_green_5 = df[(df['region'] == 5.0) & (df['golden'] >= 75)] 
df_green_6 = df[(df['region'] == 6.0) & (df['golden'] >= 75)] 
df_green_7 = df[(df['region'] == 7.0) & (df['golden'] >= 75)] 
df_green_8 = df[(df['region'] == 8.0) & (df['golden'] >= 75)] 
df_green_9 = df[(df['region'] == 9.0) & (df['golden'] >= 75)] 



#Map By Region all regions
geometry_red = [Point(xy) for xy in zip(df_red['lng'],df_red['lat'])]
geometry_yellow = [Point(xy) for xy in zip(df_yellow['lng'],df_yellow['lat'])]
geometry_green = [Point(xy) for xy in zip(df_green['lng'],df_green['lat'])]


gdf = GeoDataFrame(df_red,geometry=geometry_red)
mapa_red = gdf.plot(ax=basemap.plot(figsize=(10,6),color='#a6b8d0'), marker='o', markersize='nodos', color="red")

gdf = GeoDataFrame(df_yellow,geometry=geometry_yellow)
mapa_yellow = gdf.plot(ax=mapa_red, marker='o', markersize='nodos', color="yellow")

gdf = GeoDataFrame(df_green,geometry=geometry_green)
mapa_green = gdf.plot(ax=mapa_yellow, marker='o', markersize='nodos', color="green")
```
